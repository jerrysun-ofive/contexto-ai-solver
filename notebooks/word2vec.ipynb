### Preparation

##### tokenisation

```python
# text8 is already cleaned so no need for further preprocessing other than tokenisation
with open("../data/raw/text8.txt", "r") as f:
    raw_file = f.read()

tokens = raw_file.split()
```

```python
# check we got our tokens
print(tokens[0:10])
```

##### Vocabulary

```python
from collections import defaultdict
```

```python
# build vocabulary
token_frequency = defaultdict(lambda: 0)

for token in tokens:
    token_frequency[token] += 1
```

```python
# sanity check
print(token_frequency["use"])
```

```python
vocab = {}
vocab['<unknown>'] = 0

# pruning: remove rare words and replace with <uknown> token instead
MIN_COUNT = 5

next_id = 1
for token, count in token_frequency.items():
    if count <= MIN_COUNT:
        vocab['<unknown>'] += count
    else:
        vocab[token] = next_id

    next_id += 1
```
